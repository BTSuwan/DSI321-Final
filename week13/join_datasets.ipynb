```python
import pandas as pd

# load the first time series dataset from a CSV file
df1 = pd.read_csv('2023-05-03.csv', parse_dates=['retrived_at'])
df1.set_index('retrived_at', inplace=True)

# load the second time series dataset from a CSV file
df2 = pd.read_csv('2023-05-04.csv', parse_dates=['retrived_at'])
df2.set_index('retrived_at', inplace=True)

df3 = pd.read_csv('2023-05-05.csv', parse_dates=['retrived_at'])
df3.set_index('retrived_at', inplace=True)

# merge the two datasets based on the time index
df_merged = pd.merge(df1, df2, left_index=True, right_index=True, how='outer')


# save the merged dataset to a CSV file
df_merged.to_csv('merged_time_series.csv')
```

```python
import pandas as pd

# Load CSV files into dataframes
df1 = pd.read_csv('2023-05-03.csv')
df2 = pd.read_csv('2023-05-04.csv')
df3 = pd.read_csv('2023-05-05.csv')

# Append dataframes
appended_df = pd.concat([df1, df2, df3], ignore_index=True)

# Save appended dataframe to a new CSV file
appended_df.to_csv('appended_file.csv', index=False)
```

```python
import pandas as pd

# Load the data from the CSV file
df = pd.read_csv('3-5.csv')

# Convert the retrieved_at column to a datetime object
df['retrieved_at'] = pd.to_datetime(df['retrived_at'])

# Filter the data to only include dates between 2023-05-03 and 2023-05-05
start_date = pd.to_datetime('2023-05-03')
end_date = pd.to_datetime('2023-05-05')
mask = (df['retrieved_at'] >= start_date) & (df['retrieved_at'] <= end_date)
filtered_df = df.loc[mask]

# Calculate the average AQI for each city
city_aqi = filtered_df.groupby('city')['pollution_aqi'].mean().reset_index()
city_aqi.columns = ['city', 'average_aqi']

# Merge the average AQI data back into the original dataframe
merged_df = pd.merge(df, city_aqi, on='city', how='left')

# View the updated dataframe
print(merged_df.head())
```
